```python
import pandas as pd

import matplotlib.pyplot as plt

# ---- New dataset with different subjects & marks ----

data = {

    "Subject": ["History", "Geography", "Economics", "Computer Science", "Art"],

    "Average_Marks": [68, 72, 59, 85, 77]

}

df = pd.DataFrame(data)

# ---- Plot bar chart ----

plt.figure(figsize=(8,5))

plt.bar(df["Subject"], df["Average_Marks"], color=["teal","orange","blue","green","red"])

plt.xlabel("Subjects")

plt.ylabel("Average Marks")

plt.title("Average Marks of Students by Subject")

plt.show()

# ---- Insights ----

toughest = df.loc[df["Average_Marks"].idxmin(), "Subject"]

easiest = df.loc[df["Average_Marks"].idxmax(), "Subject"]

print(f"Toughest Subject: {toughest}")

print(f"Easiest Subject: {easiest}")

```

```python
import pandas as pd

import matplotlib.pyplot as plt

# ---- New dataset with different categories & sales ----

data = {

    "Category": ["Sports", "Toys", "Beauty", "Stationery", "Home Appliances"],

    "Sales": [25000, 18000, 32000, 15000, 40000]

}

df = pd.DataFrame(data)

# ---- Plot bar chart ----

plt.figure(figsize=(8,5))

plt.bar(df["Category"], df["Sales"], color=["purple","yellow","pink","cyan","brown"])

plt.xlabel("Product Category")

plt.ylabel("Sales")

plt.title("Sales by Product Category")

plt.show()

# ---- Insights ----

best_category = df.loc[df["Sales"].idxmax(), "Category"]

print(f"Best-Selling Category: {best_category}")
```
